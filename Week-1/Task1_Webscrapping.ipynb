```python
from bs4 import BeautifulSoup
```

```python
website="https://ackodrive.com/cars"
```

```python
response=requests.get(website)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
results = soup.find_all('div', class_="styles__ModelStatusWrapper-sc-4db1e074-14 brdlRa")
```

```python
len(results)
```

```python
results[1].find('h2').get_text()
```

```python
car_models=[]
for i in range(len(results)):
    car_model=results[i].find('h2').get_text()
    car_models.append(car_model)
car_models
```

```python
Price = soup.find_all('div', {"class": "styles__VariantPrice-sc-a6403e05-15 grIRDI"})
```

```python
Price[0].find('div',{'data-testid':"car_price"}).get_text()
```

```python
car_prices=[]
for i in range(len(Price)):
    car_price=Price[i].find('div',{'data-testid':"car_price"}).get_text()
    car_prices.append(car_price)
car_prices
```

```python
VarientInfo = soup.find_all('div', {"class": "styles__VariantInfo-sc-a6403e05-12 bkhKsh"})
fuel_type = VarientInfo[0].find('p', {'data-testid': "car_variant_fuel_type"}).get_text()
print(fuel_type)
```

```python
fuel_types = []
for i in range(len(VarientInfo)):
    fuel_type = VarientInfo[i].find('p', {'data-testid': "car_variant_fuel_type"}).get_text()
    fuel_types.append(fuel_type)

fuel_types
```

```python
Transmission=soup.find_all('div',{"class":"styles__VariantInfo-sc-a6403e05-12 bkhKsh"})
Transmission[0].find('p',{'data-testid':"car_variant_transmission"}).get_text()
```

```python
transmission_types=[]
for i in range(len(Transmission)):
    transmission_type=Transmission[i].find('p',{'data-testid':"car_variant_transmission"}).get_text()
    transmission_types.append(transmission_type)
transmission_types
```

```python
Maker=soup.find_all('div',{'class':"styles__ModelStatusWrapper-sc-4db1e074-14 brdlRa"})
Maker[0].find('span',{'class':"styles__Make-sc-a6403e05-5 etWSJY"}).get_text()
```

```python
car_makers=[]
for i in range(len(Maker)):
    car_maker=Maker[i].find('span',{'class':"styles__Make-sc-a6403e05-5 etWSJY"}).get_text()
    car_makers.append(car_maker)
car_makers
```

```python
data=[car_makers,car_makers,car_models,car_prices,fuel_types,transmission_types]
```

```python
df = pd.DataFrame(
	np.array([car_makers, car_models, car_prices, fuel_types, transmission_types]).T,
	columns=["car_makers", "car_models", "car_prices", "fuel_types", "transmission_types"]
)
```
